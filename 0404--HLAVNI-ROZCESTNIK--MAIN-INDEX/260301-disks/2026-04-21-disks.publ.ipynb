```polyglot-notebook

# i have the file "C:\Users\milan_rxtipwr\OneDrive\0-MARTI-READONLY\RECOVERY-drive-WIN-q\EFI\Boot\bootx64.efi"
# how can i get the date of expiration of the digital signature of the file ?
Get-AuthenticodeSignature "C:\Users\milan_rxtipwr\OneDrive\0-MARTI-READONLY\RECOVERY-drive-WIN-q\EFI\Boot\bootx64.efi" | Select-Object -Property Status, SignerCertificate, SignerCertificate.NotAfter
# but in winRE, there is no powershell, only cmd.exe
# so i need to get the date of expiration of the digital signature of the file in cmd.exe
# i can use the following command in cmd.exe to get the date of expiration of the digital signature of the file
certutil -dump "C:\Users\milan_rxtipwr\OneDrive\0-MARTI-READONLY\RECOVERY-drive-WIN-q\EFI\Boot\bootx64.efi" | findstr /C:"NotAfter"
#certutil -dump "C:\Users\milan_rxtipwr\OneDrive\0-MARTI-READONLY\RECOVERY-drive-WIN-q\EFI\Boot\bootx64.efi" #| findstr /C:"NotAfter"
```

```polyglot-notebook
# so which year does the digital signature of the file expire ? which year i will not be able to boot from the recovery drive ?

#!pwsh
$path = "C:\Users\milan_rxtipwr\OneDrive\0-MARTI-READONLY\RECOVERY-drive-WIN-q\EFI\Boot\bootx64.efi"
$sig = Get-AuthenticodeSignature $path

'Signer certificate validity:'
$sig.SignerCertificate | Select-Object Subject, NotBefore, NotAfter

'Timestamp certificate validity (if present):'
$sig.TimeStamperCertificate | Select-Object Subject, NotBefore, NotAfter

#$sig.

"Status: $($sig.Status)"
```
